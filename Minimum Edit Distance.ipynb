```python
import numpy as np
```

```python
#code reference: https://stackabuse.com/levenshtein-distance-and-text-similarity-in-python/

def levenshtein(s1, s2):
    x_axis = len(s1) + 1
    y_axis = len(s2) + 1
    
    matrix = np.zeros((x_axis, y_axis))
    
    for i in range(x_axis):
        matrix[i, 0] = i
    for i in range(y_axis):
        matrix[0, i] = i
        
    for i in range(1, x_axis):
        for j in range(1, y_axis):
            if s1[i-1] == s2[j-1]:
                matrix[i, j] = min(matrix[i-1, j]+1,
                                   matrix[i, j-1]+1,
                                   matrix[i-1, j-1])
            else:
                matrix[i, j] =  min(matrix[i-1, j]+1,
                                    matrix[i, j-1]+1,
                                    matrix[i-1, j-1]+2)
    print (matrix)
    print ('The min distance between %s' %s1 + ', %s is'%s2, matrix[x_axis-1, y_axis-1] )
```

```python
levenshtein('abbbbb', 'abc')
```

```python
levenshtein('intention', 'execution')
```
